```python
import pandas as pd

# Function to apply IQR-based outlier treatment
def treat_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lc = Q1 - 1.5 * IQR  # Lower cutoff
        uc = Q3 + 1.5 * IQR  # Upper cutoff

        # Cap outliers
        df[col] = df[col].clip(lower=lc, upper=uc)
    return df

# List of columns to treat
outlier_cols = ['Prothrombin','Platelets','Copper','Bilirubin']

# Apply treatment
data_encoded = treat_outliers(data_encoded, outlier_cols)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_variable_transformations(df, bins=30):
    """
    For each numeric variable, plot distributions of multiple transformations:
    - Original
    - Square
    - Square Root
    - Cube
    - Log
    Each variable will be shown in a 2x3 grid of subplots with a light peach theme.
    """
    peach_bg = "#FFE5B4"
    plot_bg = "#FFF5E1"
    text_color = "#4B3B2A"
    bar_color = "#D2691E"

    sns.set_style("whitegrid")

    transformations = {
        "Original": lambda x: x,
        "Square": lambda x: x**2,
        "Sqrt": lambda x: np.sqrt(np.clip(x, 0, None)),
        "Cube": lambda x: x**3,
        "Log": lambda x: np.log1p(np.clip(x, 0, None))
    }

    num_cols = df.select_dtypes(include=["float64", "int32", "int64"]).columns

    for col in num_cols:
        fig, axes = plt.subplots(2, 3, figsize=(18, 8))
        fig.patch.set_facecolor(peach_bg)
        fig.suptitle(f"Transformations of {col}", fontsize=16, fontweight="bold", color=text_color)

        axes = axes.flatten()
        for i, (name, func) in enumerate(transformations.items()):
            transformed = func(df[col].dropna())
            sns.histplot(transformed, kde=True, bins=bins,
                         color=bar_color, edgecolor="white", ax=axes[i])
            axes[i].set_title(name, fontsize=12, fontweight="bold", color=text_color)
            axes[i].set_xlabel("", fontsize=10, color=text_color)
            axes[i].set_ylabel("Density", fontsize=10, color=text_color)
            axes[i].tick_params(colors=text_color)
            axes[i].set_facecolor(plot_bg)
            axes[i].grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

        # Hide unused subplot if any
        if len(transformations) < len(axes):
            for j in range(len(transformations), len(axes)):
                fig.delaxes(axes[j])

        plt.tight_layout(rect=[0, 0, 1, 0.95])  # leave space for title
        plt.show()
```
